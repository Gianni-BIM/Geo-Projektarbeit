+ save the result

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# for cell independence load df from csv
df_transform_final_SHI = pd.read_csv("../output/df_transform_5.csv")

weights = {
    'pH_score': 0.05,
    'EC_score': 0.05,
    'Clay_SOC_score': 0.10,
    'P_score': 0.10,
    'N_score': 0.10,
    'K_score': 0.10,
    'erosion_score': 0.10,
    'bulk_density_score': 0.15,
    'OC_score': 0.25
    
}

# Validate weights, it it fails: Python Error
# abs(...) → ignores tiny floating-point errors
# < 1e-9 → allows extremely small numerical noise
assert abs(sum(weights.values()) - 1) < 1e-9, "Weights must sum to 1"

# function to calculate SHI (multiply each weight with corresponding column and add up all products)
def calc_shi(df, weights):
    return round(sum(df[col] * weight for col, weight in weights.items()),2)

# call the function
df_transform_final_SHI["SHI"] = calc_shi(df_transform_final_SHI, weights)

# save to final CSV, this is now the date with an SHI for each row/soil point
df_transform_final_SHI.to_csv("../output/df_transform_final_SHI.csv", index=False)
```

```python
# Check how the SHI is spread
size = len(df_transform_final_SHI)
plt.figure(figsize=(8,5))

plt.hist(df_transform_final_SHI["SHI"], bins=25, edgecolor='black', color='seagreen')

plt.xlabel("SHI")
plt.ylabel("Frequency")
plt.title(f"Distribution of Soil Health Index (SHI) in LUCAS SOIL data\nSize of dataset: {size}")

plt.show()
```
