```python
# load the merged dataset into a dataframe
# it contains LUCAS soil data: basic measurements, bulk density and texture
df_all_data = pd.read_csv("../output/df_bd_basic_texture_merged.csv")

# create an output text file for the results of data exploration
output_file = "../output/data_exploration_summary.txt"
```

### Checking '< LOD' occurences in the dataset

```python
# count the total occurence of '< LOD' values in the merged dataframe
def sum_mask_numpy(df):
    return (df.values == '< LOD').sum()

sum = sum_mask_numpy(df_all_data)
# print("=== Number of soil data points including basic measurements, bulk density and texture ===")
# print(len(df_all_data))
# print("=== Occurence of '< LOD' values in this dataset ===")
# print(sum)

with open(output_file, "w") as f:
    f.write("LUCAS Soil 2018 - Data Exploration Summary \n")
    f.write("'< LOD' values \n")
    f.write("==========================================\n\n")
    f.write("=== Number of soil data points including basic measurements, bulk density and texture ===\n")
    f.write(str(len(df_all_data)) + '\n\n')
    f.write("=== Occurence of '< LOD' values in this dataset ===\n")
    f.write(str(sum) + '\n\n')
```

```python
# find out which columns contain '< LOD' and how many times
under_LOD_counts = (df_all_data == '< LOD').sum()
lod_counts = under_LOD_counts[under_LOD_counts > 0]

# print("=== '< LOD' counts per column ===")
# print(lod_counts)

with open(output_file, "a") as f:
    f.write("=== '< LOD' counts per column ===\n")
    f.write(str(lod_counts) + '\n\n')

# 5 columns contain '< LOD' (only OC, P, N, K are needed by us)
```

```python
# the 4 relevant columns
cols = ['OC', 'P', 'N', 'K']

# count '< LOD' per row
lod_per_row = (df_all_data[cols] == '< LOD').sum(axis=1)

# summary of the counts
summary = lod_per_row.value_counts().sort_index()

# print("=== Number of '< LOD' values per row ===")
# for n_lod, n_rows in summary.items():
#     print(f"{n_lod} '< LOD' value(s)  ->  {n_rows} rows")

with open(output_file, "a") as f:
    f.write("=== Number of '< LOD' values per row ===\n")
    for n_lod, n_rows in summary.items():
        line = f"{n_rows} rows contain exactly {n_lod} '< LOD>' value(s)\n"
        f.write(line)
```

### Checking NaN or ' ' (empty fields) in the dataset

```python
# count the total occurence of '< LOD' values in the merged dataframe
columns_of_interest = ['Depth', 'BD 0-20', 'pH_CaCl2', 'EC', 'OC', 'P', 'N', 'K', 'Clay']

# print((df_all_data.isna()).sum().sum())
# print(df_all_data.isna().sum())
# same result as this expression:
# print((df_all_data.replace(r'^\s*$', None, regex=True).isna()).sum())

# check the missing values per column only for the columns of interest
# r'^\s*$' means:
# ^ start of string
# \s* any number of whitespace characters (spaces, tabs, etc.)
# $ end of string
# So it matches:
# '' (empty string)
# ' ' (only spaces)
# '\t' (tabs)
# These all become None → treated as missing (NaN)


# sum of missing values across all columns of interest
total_missing = (
    df_all_data[columns_of_interest]
    .replace(r'^\s*$', None, regex=True)
    .isna()
    .sum()
    .sum()
)
print(total_missing)


missing = (
    df_all_data[columns_of_interest]
    .replace(r'^\s*$', None, regex=True)
    .isna()
    .sum()
)
print(missing)


with open(output_file, "a") as f:
    f.write("\n")
    f.write("None/NaN values \n")
    f.write("==========================================\n\n")
    f.write("Columns of interest:\n")
    f.write(str(columns_of_interest))
    f.write("\n\n")
    f.write("=== Total of None/NaN values across columns of interst ===\n")
    f.write(str(total_missing))
    f.write("\n\n")
    f.write("=== None/NaN values per column of interst ===\n")
    f.write(missing.to_string())
    f.write("\n")
```
