# Auto-Run Data Preparation

This sript is to run all data-prep notebooks in the correct order automatically.

```python
from pathlib import Path
from nbclient import NotebookClient
import nbformat
```

```python
# Use current working directory (where you started the notebook)
BASE = Path.cwd()  

NOTEBOOKS = [
    "explore_indicator_data.ipynb",
    "data_clean_and_transform.ipynb",
    "calculate_shi.ipynb",
    "additional_expl.ipynb"
]

for nb_name in NOTEBOOKS:
    nb_path = BASE / nb_name
    print(f"\n=== Running {nb_path.name} ===")

    with open(nb_path) as f:
        nb = nbformat.read(f, as_version=4)

    client = NotebookClient(nb, timeout=600)  # 10 minutes per notebook
    client.execute()
    print(f"✅ {nb_path.name} finished")
```
